```python
'''
pub_hcb.py
Tufts Create®3 Educational Robot Example
by Courtney Kreitzer
In this example we will publish the song hot cross buns to the Create®3.
'''
import rclpy
from rclpy.node import Node
'''
These statements import messages and actions.
'''
from irobot_create_msgs.msg import AudioNote
from irobot_create_msgs.msg import AudioNoteVector
from builtin_interfaces.msg import Duration
import time

namespace = '/[Namespace]'

class SoundPublisher(Node):
    '''
    The SoundPublisher class is created which is a subclass of Node.
    This defines the class' constructor.
    '''
    def __init__(self):
        '''
        The following line calls the Node class's constructor and gives it the Node name,
        which is 'sound_publisher.'
        '''
        super().__init__('sound_publisher')
        '''
        declaring how we want to publish to the Create, in this case, we are publishing an AudioNoteVector
        '''
        self.audio_publisher_ = self.create_publisher(AudioNoteVector, namespace +
        '/cmd_audio', 10)
        '''
        The timer allows the callback to execute every 3 seconds, with a counter iniitialized.
        '''
        timer_period = 3
        self.timer = self.create_timer(timer_period, self.timer_callback)
        self.audio = AudioNoteVector()
        self.audio.append = True
    def timer_callback(self):
    	'''
    	In this function we have hardcoded the frequency and duration of notes we want the Create to play.
    	This also could have been done in a separate class
    	'''
    	self.time = Duration(sec = 1, nanosec = 0)
    	self.audio.notes = [AudioNote(frequency = 493, max_runtime = self.time), AudioNote(frequency = 440, max_runtime = self.time), AudioNote(frequency = 392, max_runtime = self.time), AudioNote(frequency = 493, max_runtime = self.time), AudioNote(frequency = 440, max_runtime = self.time), AudioNote(frequency = 392, max_runtime = self.time), AudioNote(frequency = 392, max_runtime = Duration(sec = 0, nanosec = 500000000)), AudioNote(frequency = 392, max_runtime = Duration(sec = 0, nanosec = 500000000)), AudioNote(frequency = 392, max_runtime = Duration(sec = 0, nanosec = 500000000)), AudioNote(frequency = 392, max_runtime = Duration(sec = 0, nanosec = 500000000)), AudioNote(frequency = 440, max_runtime = Duration(sec = 0, nanosec = 500000000)), AudioNote(frequency = 440, max_runtime = Duration(sec = 0, nanosec = 500000000)), AudioNote(frequency = 440, max_runtime = Duration(sec = 0, nanosec = 500000000)), AudioNote(frequency = 440, max_runtime = Duration(sec = 0, nanosec = 500000000)), AudioNote(frequency = 493, max_runtime = self.time), AudioNote(frequency = 440, max_runtime = self.time), AudioNote(frequency = 392, max_runtime = self.time)]
    	#line that publishes the audio vector to the Create
    	self.audio_publisher_.publish(self.audio)
def main(args=None):
    '''
    The rclpy library is initialized.
    '''
    rclpy.init(args=args)
    '''
    The node is created and can be used in other parts of the script.
    '''
    sound_publisher = SoundPublisher()
    try:
    	rclpy.spin_once(sound_publisher)
    except KeyboardInterrupt:
    # Destroy the node explicitly
    # (optional - otherwise it will be done automatically
    # when the garbage collector destroys the node object)
    	sound_publisher.destroy_node()
    	rclpy.shutdown()
if __name__ == '__main__':
    main()
```
